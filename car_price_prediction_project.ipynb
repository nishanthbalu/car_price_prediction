##### One-Hot encoding the categorical parameters using get_dummies

```python
dummy = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
df_add =pd.concat([df,dummy],axis =1)
df_add
```

##### Dropping relevant columns

```python
df_add = df_add.drop(['Location','Fuel_Type','Transmission','Owner_Type','Unnamed: 0','Name','New_Price'],axis=1)
df_add
```

```python
df_add
```

##### Data cleaning - Manual Replacement

```python
df_add['Mileage']=df_add['Mileage'].str.replace('km/kg','')
df_add['Mileage']=df_add['Mileage'].str.replace('kmpl','')
df_add['Mileage']=df_add['Mileage'].str.replace('null','0')

df_add['Engine']=df_add['Engine'].str.replace('CC','')
df_add['Engine']=df_add['Engine'].str.replace('null','0')

df_add['Power']=df_add['Power'].str.replace('bhp','')
df_add['Power']=df_add['Power'].str.replace('null','0')


```

```python
df_add
```

```python
df_add.dtypes
```

```python
df_add['Mileage']=df_add['Mileage'].astype(float)
df_add['Engine']=df_add['Engine'].astype(float)
df_add['Power']=df_add['Power'].astype(float)
```

```python
df_add.dtypes
```

```python
df_add.isnull().sum()
```

```python
# 0 ==> NaN
df_add.loc[df_add.Engine==0,'Engine']=np.NaN
df_add.loc[df_add.Power==0,'Power']=np.NaN
df_add.loc[df_add.Mileage==0,'Mileage']=np.NaN
```

```python
df_add.isnull().sum()
```

##### Data cleaning -Mean/Median/Mode Replacement

```python
df_add['Engine']  = df_add['Engine'].fillna(df_add['Engine'].mean())
df_add['Power']   = df_add['Power'].fillna(df_add['Power'].mean())
df_add['Mileage'] = df_add['Mileage'].fillna(df_add['Mileage'].mean())
df_add['Seats']   = df_add['Seats'].fillna(df_add['Seats'].mode()[0])
```

```python
df_add.isnull().sum()
```
